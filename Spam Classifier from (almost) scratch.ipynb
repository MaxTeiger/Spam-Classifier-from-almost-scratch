Here, we import useful libraries : 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import operator
%matplotlib inline
```

## Retrieves informations from .txt file

First, we have to put all datas from the text file to a DataFrame, which make all operations easier

```python
#open the file containing all datas
messages = open('data\\messages.txt')
messages_list = messages.readlines()
print(messages_list)
```

Now we have to create a DataFrame

```python
status, message = list(), list()

for info in messages_list:
    info.replace('\n','')
    status.append(info.split('\t')[0])
    message.append(info.split('\t')[1])

print(status)
```

Create the DataFrame from the two lists

```python
dict = {'Status':status,'Message':message}
df = pd.DataFrame(data = dict)
df.head()
```

To observe data that we have...

```python
df['Status'].value_counts()
```

Before splitting data into training and testing set, we map the values of spam and ham has 0 for ham and 1 for spam

```python
df['Labels'] = df['Status'].map({'ham': 0, 'spam': 1})
df.drop(['Status'], axis = 1, inplace = True)
df.head(100)
```

## Now we can split the data into a training and and testing set

```python
trainIndex, testIndex = list(), list()
for i in range(df.shape[0]):
    if np.random.uniform(0, 1) < 0.75:
        trainIndex += [i]
    else:
        testIndex += [i]
        
trainData = df.loc[trainIndex]
testData = df.loc[testIndex]
trainData.reset_index(inplace = True)
trainData.drop(['index'], axis = 1, inplace = True)
trainData.head()
```

Display some information to be sure the splitting is done correctly

```python
print(trainData['Labels'].value_counts(), '\n')
print(testData['Labels'].value_counts())
```

To parse both training and testing set, we can directly parse all messages. To do this : 

```python
#Create a simgle string containing all the "bad words" from ALL
#messages (not only training set, but both)

badWords = ' '.join(list(df[df['Labels'] == 1]['Message']))
print(badWords)

goodWords = ' '.join(list(df[df['Labels'] == 0]['Message']))
print(goodWords)
```

```python
badWords = badWords.split()
dictionary_bad = Counter(badWords)
goodWords = goodWords.split()
dictionary_good = Counter(goodWords)

sorted_x = sorted(dictionary_bad.items(), key=operator.itemgetter(1), reverse = True)
sorted_x
```

Now, we have two dictionaries, one containing words in spam messages as keys and their frequency as value, and the other containing words in ham messages as keys and their frequency as value
